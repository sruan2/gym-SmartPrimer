```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
baseline_per = []
baseline_imp = []
for i in range(1,16):
    pickle_name_per = 'pickles/per_baseline_'+str(i)+'.pickle'
    with open(pickle_name_per , 'rb') as handle:
        one_baseline_per = pickle.load(handle)
    baseline_per.append(one_baseline_per)
    pickle_name_imp = 'pickles/imp_baseline_'+str(i)+'.pickle'
    with open(pickle_name_imp , 'rb') as handle:
        one_baseline_imp = pickle.load(handle)
    baseline_imp.append(one_baseline_imp)
baseline_per = np.array(baseline_per)
baseline_imp = np.array(baseline_imp)
baseline_per_mean = np.mean(baseline_per, axis = 0) 
baseline_per_std = np.std(baseline_per, axis = 0) 
baseline_imp_mean = np.mean(baseline_imp, axis = 0) 
baseline_imp_std = np.std(baseline_imp, axis = 0) 
baseline_per_se = baseline_per_std/np.sqrt(15)
baseline_imp_se = baseline_imp_std/np.sqrt(15)


baseline_per_pen = []
baseline_imp_pen = []
for i in range(1,16):
    pickle_name_per_pen = 'pickles/per_baseline_pen_'+str(i)+'.pickle'
    with open(pickle_name_per_pen , 'rb') as handle:
        one_baseline_per_pen = pickle.load(handle)
    baseline_per_pen.append(one_baseline_per_pen)
    pickle_name_imp_pen = 'pickles/imp_baseline_pen_'+str(i)+'.pickle'
    with open(pickle_name_imp_pen , 'rb') as handle:
        one_baseline_imp_pen = pickle.load(handle)
    baseline_imp_pen.append(one_baseline_imp_pen)
baseline_per_pen = np.array(baseline_per_pen)
baseline_imp_pen = np.array(baseline_imp_pen)
baseline_per_pen_mean = np.mean(baseline_per_pen, axis = 0) 
baseline_per_pen_std = np.std(baseline_per_pen, axis = 0) 

baseline_imp_pen_mean = np.mean(baseline_imp_pen, axis = 0) 
baseline_imp_pen_std = np.std(baseline_imp_pen, axis = 0) 
baseline_per_pen_se = baseline_per_pen_std/np.sqrt(15)
baseline_imp_pen_se = baseline_imp_pen_std/np.sqrt(15)



```

```python

alwayshint_per = []
alwayshint_imp = []
for i in range(1,16):
    pickle_name_per = 'pickles/per_alwayshint_'+str(i)+'.pickle'
    with open(pickle_name_per , 'rb') as handle:
        one_alwayshint_per = pickle.load(handle)
    alwayshint_per.append(one_alwayshint_per)
    pickle_name_imp = 'pickles/imp_alwayshint_'+str(i)+'.pickle'
    with open(pickle_name_imp , 'rb') as handle:
        one_alwayshint_imp = pickle.load(handle)
    alwayshint_imp.append(one_alwayshint_imp)
alwayshint_per = np.array(alwayshint_per)
alwayshint_imp = np.array(alwayshint_imp)
alwayshint_per_mean = np.mean(alwayshint_per, axis = 0) 
alwayshint_per_std = np.std(alwayshint_per, axis = 0) 
alwayshint_imp_mean = np.mean(alwayshint_imp, axis = 0) 
alwayshint_imp_std = np.std(alwayshint_imp, axis = 0)
alwayshint_per_se = alwayshint_per_std/np.sqrt(15)
alwayshint_imp_se = alwayshint_imp_std/np.sqrt(15)

alwayshint_per_pen = []
alwayshint_imp_pen = []
for i in range(1,16):
    pickle_name_per_pen = 'pickles/per_alwayshint_pen_'+str(i)+'.pickle'
    with open(pickle_name_per_pen , 'rb') as handle:
        one_alwayshint_per_pen = pickle.load(handle)
    alwayshint_per_pen.append(one_alwayshint_per_pen)
    pickle_name_imp_pen = 'pickles/imp_alwayshint_pen_'+str(i)+'.pickle'
    with open(pickle_name_imp_pen , 'rb') as handle:
        one_alwayshint_imp_pen = pickle.load(handle)
    alwayshint_imp_pen.append(one_alwayshint_imp_pen)
alwayshint_per_pen = np.array(alwayshint_per_pen)
alwayshint_imp_pen = np.array(alwayshint_imp_pen)
alwayshint_per_pen_mean = np.mean(alwayshint_per_pen, axis = 0) 
alwayshint_per_pen_std = np.std(alwayshint_per_pen, axis = 0) 
alwayshint_imp_pen_mean = np.mean(alwayshint_imp_pen, axis = 0) 
alwayshint_imp_pen_std = np.std(alwayshint_imp_pen, axis = 0) 
alwayshint_per_pen_se = alwayshint_per_pen_std/np.sqrt(15)
alwayshint_imp_pen_se = alwayshint_imp_pen_std/np.sqrt(15)

```

```python
ppo_per = []
ppo_imp = []
for i in range(1,16):
    pickle_name_per = 'pickles/per_ppo_'+str(i)+'.pickle'
    with open(pickle_name_per , 'rb') as handle:
        one_ppo_per = pickle.load(handle)
    ppo_per.append(one_ppo_per)
    pickle_name_imp = 'pickles/imp_ppo_'+str(i)+'.pickle'
    with open(pickle_name_imp , 'rb') as handle:
        one_ppo_imp = pickle.load(handle)
    ppo_imp.append(one_ppo_imp)
ppo_per = np.array(ppo_per)
ppo_imp = np.array(ppo_imp)
ppo_per_mean = np.mean(ppo_per, axis = 0) 
ppo_per_std = np.std(ppo_per, axis = 0) 
ppo_imp_mean = np.mean(ppo_imp, axis = 0) 
ppo_imp_std = np.std(ppo_imp, axis = 0) 
ppo_per_se = ppo_per_std/np.sqrt(15)
ppo_imp_se = ppo_imp_std/np.sqrt(15)

ppo_per_pen = []
ppo_imp_pen = []
for i in range(1,16):
    pickle_name_per_pen = 'pickles/per_ppo_pen_'+str(i)+'.pickle'
    with open(pickle_name_per_pen , 'rb') as handle:
        one_ppo_per_pen = pickle.load(handle)
    ppo_per_pen.append(one_ppo_per_pen)
    pickle_name_imp_pen = 'pickles/imp_ppo_pen_'+str(i)+'.pickle'
    with open(pickle_name_imp_pen , 'rb') as handle:
        one_ppo_imp_pen = pickle.load(handle)
    ppo_imp_pen.append(one_ppo_imp_pen)
ppo_per_pen = np.array(ppo_per_pen)
ppo_imp_pen = np.array(ppo_imp_pen)
ppo_per_pen_mean = np.mean(ppo_per_pen, axis = 0) 
ppo_per_pen_std = np.std(ppo_per_pen, axis = 0) 
ppo_imp_pen_mean = np.mean(ppo_imp_pen, axis = 0) 
ppo_imp_pen_std = np.std(ppo_imp_pen, axis = 0) 
ppo_per_pen_se = ppo_per_pen_std/np.sqrt(15)
ppo_imp_pen_se = ppo_imp_pen_std/np.sqrt(15)
```

```python
d = {'baseline_per_mean': baseline_per_mean,
     'baseline_per_std': baseline_per_std, 
     'baseline_per_se': baseline_per_se, 
     'baseline_imp_mean': baseline_imp_mean,
     'baseline_imp_std': baseline_imp_std,
     'baseline_imp_se': baseline_imp_se,
     'baseline_per_pen_mean': baseline_per_pen_mean,
     'baseline_per_pen_std': baseline_per_pen_std, 
     'baseline_per_pen_se': baseline_per_pen_se,
     'baseline_imp_pen_mean': baseline_imp_pen_mean,
     'baseline_imp_pen_std': baseline_imp_pen_std,
     'baseline_imp_pen_se': baseline_imp_pen_se,
     
     'alwayshint_per_mean': alwayshint_per_mean,
     'alwayshint_per_std': alwayshint_per_std, 
     'alwayshint_per_se': alwayshint_per_se, 
     'alwayshint_imp_mean': alwayshint_imp_mean,
     'alwayshint_imp_std': alwayshint_imp_std,
     'alwayshint_imp_se': alwayshint_imp_se,
     'alwayshint_per_pen_mean': alwayshint_per_pen_mean,
     'alwayshint_per_pen_std': alwayshint_per_pen_std, 
     'alwayshint_per_pen_se': alwayshint_per_pen_se,
     'alwayshint_imp_pen_mean': alwayshint_imp_pen_mean,
     'alwayshint_imp_pen_std': alwayshint_imp_pen_std,
     'alwayshint_imp_pen_se': alwayshint_imp_pen_se,
     
     'ppo_per_mean': ppo_per_mean,
     'ppo_per_std': ppo_per_std, 
     'ppo_per_se': ppo_per_se, 
     'ppo_imp_mean': ppo_imp_mean,
     'ppo_imp_std': ppo_imp_std,
     'ppo_imp_se': ppo_imp_se,
     'ppo_per_pen_mean': ppo_per_pen_mean,
     'ppo_per_pen_std': ppo_per_pen_std, 
     'ppo_per_pen_se': ppo_per_pen_se,
     'ppo_imp_pen_mean': ppo_imp_pen_mean,
     'ppo_imp_pen_std': ppo_imp_pen_std,
    'ppo_imp_pen_se': ppo_imp_pen_se}
df = pd.DataFrame(data=d)
```

```python
df=pd.read_csv('data_15runs.csv')
baseline_per_mean = df['baseline_per_mean']
baseline_per_std= df['baseline_per_std']
baseline_per_se = df['baseline_per_se'] 
baseline_imp_mean = df['baseline_imp_mean']
baseline_imp_std = df['baseline_imp_std']
baseline_imp_se = df['baseline_imp_se']
baseline_per_pen_mean = df['baseline_per_pen_mean']
baseline_per_pen_std = df['baseline_per_pen_std'] 
baseline_per_pen_se = df['baseline_per_pen_se']
baseline_imp_pen_mean= df['baseline_imp_pen_mean']
baseline_imp_pen_std = df['baseline_imp_pen_std']
baseline_imp_pen_se = df['baseline_imp_pen_se']
     
alwayshint_per_mean = df['alwayshint_per_mean']
alwayshint_per_std = df['alwayshint_per_std']
alwayshint_per_se = df['alwayshint_per_se']
alwayshint_imp_mean = df['alwayshint_imp_mean']
alwayshint_imp_std = df['alwayshint_imp_std']
alwayshint_imp_se = df['alwayshint_imp_se']
alwayshint_per_pen_mean = df['alwayshint_per_pen_mean']
alwayshint_per_pen_std = df['alwayshint_per_pen_std']
alwayshint_per_pen_se = df['alwayshint_per_pen_se']
alwayshint_imp_pen_mean = df['alwayshint_imp_pen_mean']
alwayshint_imp_pen_std = df['alwayshint_imp_pen_std']
alwayshint_imp_pen_se = df['alwayshint_imp_pen_se']
     
ppo_per_mean = df['ppo_per_mean']
ppo_per_std = df['ppo_per_std']
ppo_per_se = df['ppo_per_se']
ppo_imp_mean = df['ppo_imp_mean']
ppo_imp_std = df['ppo_imp_std']
ppo_imp_se = df['ppo_imp_se']
ppo_per_pen_mean = df['ppo_per_pen_mean']
ppo_per_pen_std = df['ppo_per_pen_std']
ppo_per_pen_se = df['ppo_per_pen_se']
ppo_imp_pen_mean = df['ppo_imp_pen_mean']
ppo_imp_pen_std = df['ppo_imp_pen_std']
ppo_imp_pen_se = df['ppo_imp_pen_se']
```

```python


fig=plt.figure(figsize=(17,8))
x = list(range(1,501))
markers, caps, bars = plt.errorbar(x,baseline_per_mean, yerr =baseline_per_std ,label='random',color ='red')
[bar.set_alpha(0.3) for bar in bars]
markers, caps, bars = plt.errorbar(x,baseline_per_pen_mean, yerr =baseline_per_pen_std ,label='random+penalization',color ='magenta')
[bar.set_alpha(0.3) for bar in bars]
markers, caps, bars = plt.errorbar(x,alwayshint_per_mean, yerr =alwayshint_per_std ,label='alwayshint',color ='green')
[bar.set_alpha(0.3) for bar in bars]
markers, caps, bars = plt.errorbar(x,alwayshint_per_pen_mean, yerr =alwayshint_per_pen_std ,label='alwayshint+penalization',color ='black')
[bar.set_alpha(0.3) for bar in bars]
markers, caps, bars = plt.errorbar(x,ppo_per_mean, yerr =ppo_per_std ,label='ppo',color ='orange')
[bar.set_alpha(0.3) for bar in bars]
markers, caps, bars = plt.errorbar(x,ppo_per_pen_mean, yerr =ppo_per_pen_std ,label='ppo+penalization',color ='blue')
[bar.set_alpha(0.3) for bar in bars]
plt.title('Average reward of last 50 children (SD)', fontsize = 22)
plt.xlabel('Episodes') 
plt.legend(prop={'size': 10})
plt.show()
```
