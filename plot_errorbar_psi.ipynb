```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
ppo_per = []
ppo_imp = []
ppo_question = []
ppo_encourage = []
ppo_hint = []
ppo_nothing = []
for i in range(1,16):
    pickle_name_per = 'pickles/per_ppo_'+str(i)+'.pickle'
    with open(pickle_name_per , 'rb') as handle:
        one_ppo_per = pickle.load(handle)
    ppo_per.append(one_ppo_per)
    pickle_name_imp = 'pickles/imp_ppo_'+str(i)+'.pickle'
    with open(pickle_name_imp , 'rb') as handle:
        one_ppo_imp = pickle.load(handle)
    ppo_imp.append(one_ppo_imp)
    
    pickle_name_actionInfo = 'pickles/actionInfo_ppo_'+str(i)+'.pickle'
    with open(pickle_name_actionInfo , 'rb') as handle:
        one_ppo_actionInfo = pickle.load(handle)
    ppo_question.append(one_ppo_actionInfo['question'][0:10104])
    ppo_encourage.append(one_ppo_actionInfo['encourage'][0:10104])
    ppo_hint.append(one_ppo_actionInfo['hint'][0:10104])
    ppo_nothing.append(one_ppo_actionInfo['nothing'][0:10104])

    
    
ppo_per = np.array(ppo_per)
ppo_imp = np.array(ppo_imp)
ppo_question = np.array(ppo_question)
ppo_encourage = np.array(ppo_encourage)
ppo_hint = np.array(ppo_hint)
ppo_nothing = np.array(ppo_nothing)

ppo_per_mean = np.mean(ppo_per, axis = 0) 
ppo_per_std = np.std(ppo_per, axis = 0) 
ppo_per_se = ppo_per_std/np.sqrt(15)

ppo_imp_mean = np.mean(ppo_imp, axis = 0) 
ppo_imp_std = np.std(ppo_imp, axis = 0) 
ppo_imp_se = ppo_imp_std/np.sqrt(15)

ppo_question_mean = np.mean(ppo_question, axis = 0) 
ppo_question_std = np.std(ppo_question, axis = 0) 
ppo_question_se = ppo_question_std/np.sqrt(15)

ppo_encourage_mean = np.mean(ppo_encourage, axis = 0) 
ppo_encourage_std = np.std(ppo_encourage, axis = 0) 
ppo_encourage_se = ppo_encourage_std/np.sqrt(15)

ppo_hint_mean = np.mean(ppo_hint, axis = 0) 
ppo_hint_std = np.std(ppo_hint, axis = 0) 
ppo_hint_se = ppo_hint_std/np.sqrt(15)

ppo_nothing_mean = np.mean(ppo_nothing, axis = 0) 
ppo_nothing_std = np.std(ppo_nothing, axis = 0) 
ppo_nothing_se = ppo_nothing_std/np.sqrt(15)
```

```python
ppo_psi03_per = []
ppo_psi03_imp = []
ppo_psi03_question = []
ppo_psi03_encourage = []
ppo_psi03_hint = []
ppo_psi03_nothing = []
for i in range(1,16):
    pickle_name_per = 'pickles/per_ppo_psi03_'+str(i)+'.pickle'
    with open(pickle_name_per , 'rb') as handle:
        one_ppo_per = pickle.load(handle)
    ppo_psi03_per.append(one_ppo_per)
    
    pickle_name_imp = 'pickles/imp_ppo_psi03_'+str(i)+'.pickle'
    with open(pickle_name_imp , 'rb') as handle:
        one_ppo_imp = pickle.load(handle)
    ppo_psi03_imp.append(one_ppo_imp)
    
    pickle_name_actionInfo = 'pickles/actionInfo_ppo_psi03_'+str(i)+'.pickle'
    with open(pickle_name_actionInfo , 'rb') as handle:
        one_ppo_actionInfo = pickle.load(handle)
    ppo_psi03_question.append(one_ppo_actionInfo['question'][0:10350])
    ppo_psi03_encourage.append(one_ppo_actionInfo['encourage'][0:10350])
    ppo_psi03_hint.append(one_ppo_actionInfo['hint'][0:10350])
    ppo_psi03_nothing.append(one_ppo_actionInfo['nothing'][0:10350])

    
    
    
ppo_psi03_per = np.array(ppo_psi03_per)
ppo_psi03_imp = np.array(ppo_psi03_imp)
ppo_psi03_question = np.array(ppo_psi03_question)
ppo_psi03_encourage = np.array(ppo_psi03_encourage)
ppo_psi03_hint = np.array(ppo_psi03_hint)
ppo_psi03_nothing = np.array(ppo_psi03_nothing)

ppo_psi03_per_mean = np.mean(ppo_psi03_per, axis = 0) 
ppo_psi03_per_std = np.std(ppo_psi03_per, axis = 0) 
ppo_psi03_per_se = ppo_psi03_per_std/np.sqrt(15)

ppo_psi03_imp_mean = np.mean(ppo_psi03_imp, axis = 0) 
ppo_psi03_imp_std = np.std(ppo_psi03_imp, axis = 0) 
ppo_psi03_imp_se = ppo_psi03_imp_std/np.sqrt(15)

ppo_psi03_question_mean = np.mean(ppo_psi03_question, axis = 0) 
ppo_psi03_question_std = np.std(ppo_psi03_question, axis = 0) 
ppo_psi03_question_se = ppo_psi03_question_std/np.sqrt(15)

ppo_psi03_encourage_mean = np.mean(ppo_psi03_encourage, axis = 0) 
ppo_psi03_encourage_std = np.std(ppo_psi03_encourage, axis = 0) 
ppo_psi03_encourage_se = ppo_psi03_encourage_std/np.sqrt(15)

ppo_psi03_hint_mean = np.mean(ppo_psi03_hint, axis = 0) 
ppo_psi03_hint_std = np.std(ppo_psi03_hint, axis = 0) 
ppo_psi03_hint_se = ppo_psi03_hint_std/np.sqrt(15)

ppo_psi03_nothing_mean = np.mean(ppo_psi03_nothing, axis = 0) 
ppo_psi03_nothing_std = np.std(ppo_psi03_nothing, axis = 0) 
ppo_psi03_nothing_se = ppo_psi03_nothing_std/np.sqrt(15)


```

```python


fig=plt.figure(figsize=(17,8))
x = list(range(1,501))

markers, caps, bars = plt.errorbar(x,ppo_per_mean, yerr =ppo_per_std ,label='ppo',color ='orange')
[bar.set_alpha(0.3) for bar in bars]
markers, caps, bars = plt.errorbar(x,ppo_psi03_per_mean, yerr =ppo_psi03_per_std ,label='ppo-psi-0.3',color ='blue')
[bar.set_alpha(0.3) for bar in bars]
plt.title('Average reward of last 50 children (SD)  16', fontsize = 22)
plt.xlabel('Episodes') 
plt.legend(prop={'size': 10})
plt.show()


fig=plt.figure(figsize=(17,8))
x = list(range(1,501))

markers, caps, bars = plt.errorbar(x,ppo_per_mean, yerr =ppo_per_se ,label='ppo',color ='orange')
[bar.set_alpha(0.3) for bar in bars]
markers, caps, bars = plt.errorbar(x,ppo_psi03_per_mean, yerr =ppo_psi03_per_se ,label='ppo-psi-0.3',color ='blue')
[bar.set_alpha(0.3) for bar in bars]
plt.title('Average reward of last 50 children (SE)  16', fontsize = 22)
plt.xlabel('Episodes') 
plt.legend(prop={'size': 10})
plt.show()
```
