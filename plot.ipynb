```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('data.csv')
baseline = df['baseline']
baseline_penalization = df['baseline_penalization']
alwayshint = df['alwayshint']
alwayshint_penalization = df['alwayshint_penalization']
ppo = df['ppo']
ppo_penalization = df['ppo_penalization']
```

```python
fig=plt.figure(figsize=(17,8))
plt.plot(baseline,label='random',color ='red')
plt.plot(baseline_penalization,label='random+penalization',color ='magenta')
plt.plot(alwayshint,label='alwayshint',color ='green')
plt.plot(alwayshint_penalization,label='alwayshint+penalization',color ='cyan')
plt.plot(ppo,label='ppo',color ='orange')
plt.plot(ppo_penalization,label='ppo+penalization',color ='blue')
plt.title('Average reward of last 50 children', fontsize = 22)
plt.xlabel('Epochs') 
plt.legend(prop={'size': 10})
plt.show()
```
